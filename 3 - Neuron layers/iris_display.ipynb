# Display of Iris data

```python
from sklearn import datasets
from sklearn.manifold import TSNE

import numpy as np
import pylab as plt


seed = 50
np.random.seed(seed)
```

## Read Iris data 

```python
iris = datasets.load_iris()
iris.data -= np.mean(iris.data, axis=0)

X, Y = iris.data, iris.target

no_data = len(X)
idx = np.arange(no_data)
np.random.shuffle(idx)
X, Y = X[idx], Y[idx]

print('no of data: %d'%no_data)
print(X[:10])
print(Y[:10])
```

```python
# plot the decision boundary in 2D spaces

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10,10))
fig.suptitle('boundaries in 2-dimensional feature spaces')

x1 = np.arange(min(X[:,0]), max(X[:,0]), 0.1)
x2 = np.arange(min(X[:,1]), max(X[:,1]), 0.1)
x3 = np.arange(min(X[:,2]), max(X[:,2]), 0.1)
x4 = np.arange(min(X[:,3]), max(X[:,3]), 0.1)

ax1.plot(X[Y==0,0], X[Y==0,1],'bx', label ='class 1')
ax1.plot(X[Y==1,0],X[Y==1,1],'ro', label='class 2')
ax1.plot(X[Y==2,0], X[Y==2,1],'g+', label ='class 3')
ax1.set_xlabel('$x_1$')
ax1.set_ylabel('$x_2$')
ax1.legend()

ax2.plot(X[Y==0,1], X[Y==0,2],'bx', label ='class 1')
ax2.plot(X[Y==1,1],X[Y==1,2],'ro', label='class 2')
ax2.plot(X[Y==2,1], X[Y==2,2],'g+', label ='class 3')
ax2.set_xlabel('$x_2$')
ax2.set_ylabel('$x_3$')
ax2.legend()

ax3.plot(X[Y==0,2], X[Y==0,3],'bx', label ='class 1')
ax3.plot(X[Y==1,2],X[Y==1,3],'ro', label='class 2')
ax3.plot(X[Y==2,2], X[Y==2,3],'g+', label ='class 3')
ax3.set_xlabel('$x_3$')
ax3.set_ylabel('$x_4$')
ax3.legend()

ax4.plot(X[Y==0,3], X[Y==0,0],'bx', label ='class 1')
ax4.plot(X[Y==1,3],X[Y==1,0],'ro', label='class 2')
ax4.plot(X[Y==2,3], X[Y==2,0],'g+', label ='class 3')
ax4.set_xlabel('$x_4$')
ax4.set_ylabel('$x_1$')
ax4.legend()

plt.show()
```

## Plot them in 2-D after transforming with Tsne

```python
XX = TSNE(random_state=10).fit_transform(X)

print(XX.shape)

plt.figure(2)
plt.plot(XX[Y==0,0], XX[Y==0,1], 'b+', label='Setosa')
plt.plot(XX[Y==1,0], XX[Y==1,1], 'r.', label='Versicolor')
plt.plot(XX[Y==2,0], XX[Y==2,1], 'gx', label='Virginica')
plt.xlabel('$y_1$')
plt.ylabel('$y_2$')
plt.legend()
plt.show()
```
